```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
results = pd.read_csv("/content/results.csv",sep=",")
```

```python
exclude_categories = ['micro avg', 'macro avg', 'weighted avg', 'samples avg']
filtered_df = results[~results['Category'].isin(exclude_categories)]
```

```python
unique_categories = results['Model Name'].unique()
print(unique_categories)
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Model Name', y='f1-score', data=filtered_df)
plt.title('Box Plot of F1-scores by Model Configuration')
plt.ylabel('F1 Score')
plt.xlabel('Model')


new_xticks = ["SafeAeroBERT_BCE", "Longformer-base-4096_BCE", "bert-base-uncased_Unfrozen_BCE","SafeAeroBERT_Unfrozen_BCE","Longformer-base-4096_Unfrozen_BCE","bert-base-uncased_BCE","bert-base-uncased_BCE-Balanced", "SafeAeroBERT_BCE-Balanced","Longformer-base-4096_BCE-Balanced", "bert-base-uncased_Unfrozen_BCE-Balanced","SafeAeroBERT_Unfrozen_BCE-Balanced", "Longformer-base-4096_Unfrozen_BCE-Balanced","bert-base-uncased_BinaryFocal","SafeAeroBERT_BinaryFocal", "Longformer-base-4096_BinaryFocal","bert-base-uncased_Unfrozen_BinaryFocal","SafeAeroBERT_Unfrozen_BinaryFocal"]
plt.xticks(range(len(new_xticks)), new_xticks, rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Model Name', y='binary_accuracy', data=filtered_df)
plt.title('Box Plot of Accuracy by Model Configuration')
plt.ylabel('Accuracy')
plt.xlabel('Category')


new_xticks = ["SafeAeroBERT_BCE", "Longformer-base-4096_BCE", "bert-base-uncased_Unfrozen_BCE","SafeAeroBERT_Unfrozen_BCE","Longformer-base-4096_Unfrozen_BCE","bert-base-uncased_BCE","bert-base-uncased_BCE-Balanced", "SafeAeroBERT_BCE-Balanced","Longformer-base-4096_BCE-Balanced", "bert-base-uncased_Unfrozen_BCE-Balanced","SafeAeroBERT_Unfrozen_BCE-Balanced", "Longformer-base-4096_Unfrozen_BCE-Balanced","bert-base-uncased_BinaryFocal","SafeAeroBERT_BinaryFocal", "Longformer-base-4096_BinaryFocal","bert-base-uncased_Unfrozen_BinaryFocal","SafeAeroBERT_Unfrozen_BinaryFocal"]
plt.xticks(range(len(new_xticks)), new_xticks, rotation=45, ha='right')
plt.show()
```
